## Inline Question 1: 

We've only asked you to implement ReLU, but there are a number of different activation functions that one could use in neural networks, each with its pros and cons. In particular, an issue commonly seen with activation functions is getting zero (or close to zero) gradient flow during backpropagation. Which of the following activation functions have this problem? If you consider these functions in the one dimensional case, what types of input would lead to this behaviour?
1. Sigmoid
2. ReLU
3. Leaky ReLU

## Answer:
Close to zero gradient flow during backpropagation leads to the vanishing gradient problem.

1. The Sigmoid function can lead to the vanishing gradient issue because its gradient approaches zero when the input values are extremely high or low. This is evident when we look at the tails of the function. For instance, when dealing with large one-dimensional values like [-1e3, 1e3], the function tends to saturate.

2. A significant advantage of the ReLU (Rectified Linear Unit) over the Sigmoid function is its resilience against the vanishing gradient problem, especially when handling positive inputs. Specifically, the gradient of ReLU is either 0 (for negative inputs) or 1 (for positive ones). However, ReLU isn't immune to problems. If all input values are negative—though this is an uncommon scenario—the gradient can vanish. This phenomenon is termed the "dying ReLU problem." To illustrate this, consider a one-dimensional input with only negative values like [-1, -2, -3].

3. Leaky ReLU aims to tackle the issue of non-active neurons that ReLU sometimes encounters. It does this by introducing a minor negative slope for negative inputs, following the formula: if x < 0, then 0.01x, otherwise x. This makes Leaky ReLU more robust against the vanishing gradient problem. Nonetheless, a potential challenge arises from the max(0.01x, x) function being non-continuous at x = 0, leading to an undefined gradient at that point. Unless addressed in the coding, a scenario that might cause zero gradients is when all input values are zeroes, for instance [0, 0, 0]. This can be a consequence of improper network initialization.

## Answer:
1. When contrasting the challenges of training a three-layer network with a five-layer one, the latter displayed greater susceptibility to variations in weight initialization. This heightened sensitivity stems from the five-layer network being 66% deeper. In deep networks, too small weight scales can cause the gradients to vanish (due to tiny products), while overly large ones can cause the gradients to explode (because of sizable products). The risk of vanishing gradient problems amplifies with network depth. Consequently, pinpointing the ideal weight initialization for a five-layer network proves more challenging than for its three-layer counterpart.

2. The following example highlights this sensitivity in the five-layer network concerning weight initialization. Starting with a minimal weight scale of 1e-5, both the mean and standard deviation of the parameters were nearly zero, indicative of vanishing gradients. Boosting the weight scale by several orders of magnitude brought improvements, with the optimal training performance observed at a weight scale of 5e-2. Pushing it slightly to 1e-1, however, results in exploding gradients. Comparatively, networks with fewer hidden layers, like the three-layer one, still react to weight scale adjustments, but their reactions are more muted. The ideal weight scale for the three-layer network is broader (around 2e-2), unlike the five-layer network's more precise requirement (5e-2). Additionally, it's worth noting that the learning rate also influences these weight scale values.
